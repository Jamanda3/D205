### Part II: Data-Cleaning Plan

#### C.1.  Propose a plan that includes the relevant techniques and specific steps needed to identify anomalies in the data set. 

My plan to clean the data set is to go through it thoroughly and methodically from a high-level overview down to the individual values. After examining the structure and checking for duplicate rows, go though each column evaluating the data types, range of values, and missing values. The next step is to drill down on the values to verify they are consistent with the range and format defined in the data dictionary that accompanied the data set, while simultaneously looking for hidden missing values that have been replaced with placeholders. Further evaluation will be performed on values that can be checked for accuracy and relevance. Distribution will be visualized with histograms to look for symmetry or skewedness. Columns that are inaccurate, redundant, or irrelevant will be dropped. Misleading or ambiguous column names will be renamed. The index will be reset if needed. Missing values will then be evaluated for patterns of missingness. Re-expression of the categorical variables will be done before treating the missing values if multivariant imputation is needed. Missing values will be imputed with the method that best maintains the original distribution. Outliers will be evaluated with boxplots and treated if needed.(Larose, C. D., & Larose, D. T. (2015)) 

Detailed account of the process and techniques:  
1. Import needed packages  
2. Load the data set  
3. Make a backup copy before manipulating it  
4. Examine data set as a whole  
    a. `.shape-shape of the data frame`  
    b. `.info()- column names, none-null counts, dtypes`  
    c. `.head()- see how it displays`  
5. Check for duplicate rows- `.duplicated().sum()`  
6. Examine columns  
    a. `.describe()`- numeric variables-count, mean, std, min, max  
    b. `.describe(exclude='number')`- object variables- count, unique, top  
7. Examine each variable. Identify hidden missing, unique, improper, or anomalous values  
    a. Numeric Variables- `.describe()`, `.nunique()`, `.value_counts()`  
    b. Object variables- `.value_counts()`, `.str.contains()`, `sort_values()`
8. Delete redundant and inaccurate columns- `df = df.drop(['col_name'], axis=1)`
9. Rename misleading columns- `df= df.rename(columns={'org_col': 'new_col'})`
10. Reset Index- `df.set_index('col_name', drop = True, inplace = True)`  
11. Examine missing values    
    a. Missing count per variable- `df.isnull().sum()`  
    b. Sub-set variables with nulls- `vars_with_na = df.loc[:, ('col_name1', 'col_name2')`]  
    c. Percentage of missing- `data-vars_with_na.isnull().mean()`  
12. Examine missing values found for type of missingness  
    a. `msno.matrix(df)`  
    b. `msno.heatmap(df)`  
    c. `msno.dendrogram(df)`  
13. Re-expression of categorical variables(needed before imputing for MICE)  
    a. Ordinal variables- Ordinal encoder  
    b. Nominal variables- Label encoder
14. Impute missing values- mean, mode, median, KNN, and MICE imputations were compared and method that best maintained the original distribution was used  
15. Distributions were visualized with histograms and distribution curves.  
16. Assess for outliers- boxplots  
17. Treat outliers

#### Mean, Median, Mode

```python
# Datacamp: Dealing with Missing Data in Python, Chapter 3:Imputation Techniques 
df_mean = df.copy(deep=True)
mean_imputer = SimpleImputer(strategy='mean') 
df_mean.iloc[:, :] = mean_imputer.fit_transform(df_mean) 
# Median imputer
df_median = df.copy(deep=True)
median_imputer = SimpleImputer(strategy='median')
df_median.iloc[:, :] = median_imputer.fit_transform(df_median)
# Mode Imputer
df_mode = df.copy(deep=True)
mode_imputer = SimpleImputer(strategy='most_frequent')
df_mode.iloc[:, :] = mode_imputer.fit_transform(df_mode)
```

#### KNN

```python
# Datacamp: Dealing with Missing Data in Python, Chapter 4:Advanced Imputation Techniques
df_KNN = df.copy(deep=True)  
KNN_imputer = KNN() 
df_KNN.iloc[:, :] = KNN_imputer.fit_transform(df_KNN)
```

#### MICE

```python
# https://stackoverflow.com/questions/61945250/im-getting-negative-values-as-output-of-iterativeimputer-from-sklearn
min_list = df.iloc[:,:].min(axis=0)
max_list = df.iloc[:,:].max(axis=0)
```

```python
# Datacamp: Dealing with Missing Data in Python, Chapter 4:Advanced Imputation Techniques
from fancyimpute import IterativeImputer #
MICE_imputer = IterativeImputer(min_value = min_list, max_value=max_list)
df_MICE = df.copy(deep=True) 
df_MICE.iloc[:, :] = MICE_imputer.fit_transform(df_MICE)
```

For the PCA I chose to include the variables that were most directly  related to patient health and hospital stay: 'Age', 'Gender', 'VitD_levels', 'Doc_visits', 'Full_meals_eaten', 'VitD_supp', 'Soft_drink', 'Initial_admin', 'HighBlood', 'Stroke',  'Complication_risk','Overweight', 'Arthritis', 'Diabetes', 'Hyperlipidemia', 'BackPain', 'Anxiety', 'Allergic_rhinitis', 'Reflux_esophagitis', 'Asthma', 'Services', 'Initial_days', 'TotalCharge', 'Additional_charges'. #### E.2.  Describe how you identified the principal components of the data set. 

After normalizing the data and applying PCA to it, eigenvalues were computed. Using the Kaiser rule, PCs with eigenvalues greater than one were kept. (Larose, C. D., & Larose, D. T. (2015))

#### E.3 Describe how the organization can benefit from the results of the PCA

Hospitals can use the factors identified in PCA to identify patients at risk for readmission and  develope protocols to address the factors to reduce readmission rate. This would not only reduce costs but ultimately improve patient care and outcomes.

```python
#Define variables for PCA- Selecting variables directly related to patient health and hospital stay
df_pca = df[[ 'Age', 'Gender', 'VitD_levels', 'Doc_visits', 'Full_meals_eaten', 'VitD_supp', 'Soft_drink', 'Initial_admin', 'HighBlood', 'Stroke', 
             'Complication_risk','Overweight', 'Arthritis', 'Diabetes', 'Hyperlipidemia', 'BackPain', 'Anxiety', 'Allergic_rhinitis', 
             'Reflux_esophagitis', 'Asthma', 'Services', 'Initial_days', 'TotalCharge', 'Additional_charges' 
            ]]
```

```python
#Normalize data
df_pca_normalized = (df_pca-df_pca.mean())/df_pca.std()
```

```python
# Indentify number of PCs to extract
pca = PCA(n_components=df_pca.shape[1])
```

```python
#Apply the PCA
pca.fit(df_pca_normalized)
```

```python
df_pca2 = pd.DataFrame(pca.transform(df_pca_normalized), columns= ['PC1','PC2','PC3','PC4', 'PC5','PC6','PC7','PC8','PC9','PC10', 'PC11',
                                                                   'PC12','PC13','PC14','PC15','PC16','PC17','PC18','PC19', 'PC20', 'PC21',
                                                                   'PC22','PC23','PC24'])
```

```python
loadings = pd.DataFrame(pca.components_.T,
                       columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13', 'PC14',
                                'PC15','PC16','PC17','PC18','PC19','PC20', 'PC21','PC22','PC23','PC24'],
                       index=df_pca_normalized.columns)
loadings
```

```python
cov_matrix = np.dot(df_pca_normalized.T, df_pca_normalized)/df_pca.shape[0]  #Larose, C. D., & Larose, D. T. (2015)
eigenvalues = [np.dot(eiganvector.T, np.dot(cov_matrix,eiganvector)) for eiganvector in pca.components_]
```

```python
fig = plt.figure(figsize = (10,7))
plt.plot(eigenvalues)
plt.xlabel('Number of Components')
plt.ylabel('Eigenvalues')
plt.show
```
